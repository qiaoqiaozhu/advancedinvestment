### Boolean
A boolean expression (or logical expression) evaluates to one of two states true or false. Python provides the boolean type that can be either set to False or True. Many functions and operations returns boolean objects.

For "and" case, if "$x$ and $y$" is given, returns $x$ if $x$ is False, $y$ otherwise, examples:

```python
5 > 2 and 3 > 1
```

```python
5 > 2 and 3 <= 1
```

```python
5 == 2 and 3 <= 1
```

For "or" case, if "$x$ or $y$" is given, returns $y$ if $x$ is False, $x$ otherwise, examples:

```python
5 > 2 or 3 <= 1
```

```python
5 != 5 or 3 < 1
```

For "not" case, if "not $x$" is given, returns False if $x$ is True, True otherwise, examples:

```python
not 3 > 2
```

```python
not 3 < 2
```

### Conditional Statements
In daily communication, we also make conditional statements. Consider the conversation below, for instance:



If it rains heavily tomorrow, then I will do assignments. 

If it only drizzles I'll sleep.

If it does not rain tomorrow, I'll play football.

OK, let's try to operationalize this example here in Python. Suppose I set the state *weather* to be *rain*:

```python
weather = "rain"  # rain | drizzle | no rain
```

By the way, the # sign to the right of the code above signifies what follows it to be *comments*. Comments will not be interpreted nor acted upon by Python.

Now here are the conditionals and the output, given our instance of *weather* as *rain*:

```python
if weather == "rain":
    print ("Weather = " + weather + ": do assignments") 
elif weather == "drizzle":
    print ("Weather = " + weather + ": sleep")
else:
    print ("Weather = " + weather + ": play football")
```

### Functions

```python
def CAPM(rf, beta, rm):
    ri = rf + beta * (rm - rf)
    print (ri)
```

```python
rf = 0.03
beta = 1.2
rm = 0.05
CAPM(rf, beta, rm)
```

Two things: (1) format, (2) objects within a function are defined only in the scope of the function.

Or, we can write a loop!

Here's another nifty Pythonic function, called *range*:

```python
range(20)
```

So let's use this as so:

```python
for i in range(20):
    print ("This is step " + str(i+1))
```

Observe that I used something else above: The *str* function, another Pythonic built-in function. This converts its argument into a *string*, i.e., a word, as opposed to a numeric variable. Also I used the "+" operator in the print statement in the loops. This is a different usage to the numerical addition operator. Here's Python knows to interpret "+" as the concatenation of two string variables.

#### Date Time Index

```python
df.index = pd.to_datetime(df.index.astype('str'))
```

```python
df['year'] = df.index.year
```

#### Aggregation and rgoupby operations

```python
df.groupby('year')['ex_ret'].mean()
```

```python
import statsmodels.api as sm 
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars] 
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit() 
    return result.params
```

```python
by_year = df[df.year>=2008].groupby('year')  #bollean selection 
```

```python
by_year.apply(regress, 'ex_ret', ['mktrf'])
```
