### Modeling Mean-reversion
Let $x_{t}\equiv E_{t}r_{t+1}-r_{f}+\frac{\sigma^{2}_{\epsilon}}{2}$ as expected excess log return. Let's write the return process as
$$
\begin{align} r_{t+1}=E_{t}r_{t+1}+\epsilon_{t+1}  \\
 x_{t+1}=\phi x_{t} + (1-\phi) \mu +\eta_{t+1}
\end{align}
$$

where $\phi \in (0,1)$ and error terms are iid, but $\epsilon, \eta$ might be correlated. Notice that

  * If we take expectation on (1b), we have $E(x_{t})=\mu$, unconditional expectation of the excess return is $\mu$
  * If $\phi = 0$, then the model reduces to constant expected return and random walk model. 

From the perspective of realized returns, 

$$ r_{t+1}=E_{t}r_{t+1}+\epsilon_{t+1} = x_{t}+r_{f}-\frac{\sigma^{2}_{\epsilon}}{2}+\epsilon_{t+1} \\
= \phi x_{t-1} + (1-\phi) \mu +\eta_{t}+r_{f}-\frac{\sigma^{2}_{\epsilon}}{2}+\epsilon_{t+1} \\
= \phi (E_{t-1}r_{t}-r_{f}+\frac{\sigma^{2}_{\epsilon}}{2}) + (1-\phi) \mu +\eta_{t}+r_{f}-\frac{\sigma^{2}_{\epsilon}}{2}+\epsilon_{t+1}\\
= \phi (r_{t} -\epsilon_{t} - r_{f}+\frac{\sigma^{2}_{\epsilon}}{2}) + (1-\phi) \mu +\eta_{t}+r_{f}-\frac{\sigma^{2}_{\epsilon}}{2}+\epsilon_{t+1} \\
= \phi r_{t} + (1-\phi) (\mu +r_{f}-\frac{\sigma^{2}_{\epsilon}}{2}) +\eta_{t}+\epsilon_{t+1} -\epsilon_{t}
$$

It also follows AR(1) process

```python
fit = smf.ols('excessret ~ excessret.shift()', data=gw['1953':].dropna()).fit()
print(fit.summary())
```

covariance of return process

$$cov_{t}(r_{t+1}, r_{t+2})=cov_{t}(r_{t+1}, E_{t+1}r_{t+2}+\epsilon_{t+2})=cov_{t}(r_{t+1}, x_{t+1}+r_{f}-\frac{\sigma^{2}_{\epsilon}}{2}+\epsilon_{t+2}) \\
=cov_{t}(r_{t+1}, x_{t+1}) =cov_{t}(r_{t+1}, \phi x_{t} + (1-\phi) \mu +\eta_{t+1})\\
=cov_{t}(E_{t}r_{t+1}+\epsilon_{t+1}, x_{t+1}) =cov_{t}(E_{t}r_{t+1}+\epsilon_{t+1}, \phi x_{t} + (1-\phi) \mu +\eta_{t+1})\\
=cov_{t}(\epsilon, \eta)=\sigma_{\epsilon\eta}$$

Notice that $E_{t}r_{t+1}$ is not a source of risk for investors. 

Mean reversion conditional on time t information means $\sigma_{\epsilon\eta}<0$, an unexpectedly high return today (high $\epsilon_{t}$) reduces expected returns in the future (low $\eta$).

### Long Investment Horizon 

It is often suggested stocks are safer in the long-run. long term investors can "afford to wait out the ups and
downs of the market." (e.g. Siegel's book "Stocks for the Long Run") This suggests that as investment horizon k increases, annualized volatility
$VAR_{t}(r_{t\rightarrow t+k} )/K$ decreases. We should see it is only true if returns are mean-reverse. With log return,
$$ r_{t\rightarrow t+k}=r_{t+1}+r_{t+2}+\cdots +r_{t+k} $$

#### Multi-Period Variance

When returns are iid
$$ VAR_{t}(r_{t\rightarrow t+k})=VAR_{t}(r_{t+1}+r_{t+2}+\cdots +r_{t+k}) = kVAR_{t}(r_{t+1}) $$

variance of Long-horizon return grows with horizon. If returns mean-revert then

$$\begin{align} VAR_{t}(r_{t\rightarrow t+2})&=VAR_{t}(r_{t+1})+VAR_{t}(r_{t+2})+2COV_{t}(r_{t+1},r_{t+2})\\&=2VAR_{t}(r_{t+1})+2COV_{t}(r_{t+1},r_{t+2})<2VAR_{t}(r_{t+1})\end{align}$$

Stocks appear relatively safer to long-horizon investors. This also suggest a variance-ratio test for mean reversion:

$$ v_{k}=\frac{VAR_{t}(r_{t\rightarrow t+k})}{kVAR_{t}(r_{t+1})}$$

Mean reversion in k if $v_{k}<1$.

conditional covariance of future returns are negative if expected returns follow a mean-reverting process. 

Unconditional variance can be quite different since
$$ cov(r_{t+1}, r_{t+2}) = cov(r_{t+1}, x_{t+1})=cov(x_{t}+\epsilon_{t+1}, x_{t+1})=\phi \sigma_{x}^{2}+\sigma_{\epsilon\eta}$$ 
persistence in $x_{t+1}$ can make unconditional covariance close to zero or positive even if conditional covariance is negative. In other words, return correlation can be hard to detect in the data ex post. 

### Asset Allocation

How would the investment decision affect asset allocation decision. Recall the decision of investing in risky assets from early decisions. 

Suppose you have monthly data from 1953 to 2011, with power utility function and risk aversion of 5. We can find the optimal weight in risky assets for a one period investment through simulation or optimization. 

```python
def U(w, riskaversion):
    return w**(1-riskaversion)/(1-riskaversion)

def futurewealth_logret(s, rf, ret): 
    return (1-s)*np.exp(rf) + s*np.exp(ret)

def multiperiodret_random(rf, mean, sigma, k, n):
    return rf*k , sigma * np.random.randn(n, n, k).sum(axis=2) + k*mean

def alloc_simplot(rf, mean, sigma, k, n, gamma, bins=200):
    S=np.linspace(0, 1, bins)
    expu=[]
    for s in S:
        u = U(futurewealth_logret(s, multiperiodret_random(rf, mean, sigma, k, n)[0], 
                                  multiperiodret_random(rf, mean, sigma, k, n)[1]), gamma)
        m = u.mean(axis=1).mean(axis=0)
        expu.append(m)
    sim = pd.DataFrame(np.column_stack([S, mexpu]), columns=['share', 'exputility'])
    print('Stock allocation for multiperiod Maximum Expected Utility: %2f.' %sim.loc[sim.exputility.argmax(), 'share'])
    fig = plt.figure()
    plt.plot(sim.share, sim.exputility)
    plt.xlabel('Allocation to Stocks')
    plt.ylabel('Expected Utility')
    plt.show()


def alloc_sim(rf, mean, sigma, k, n, gamma, bins=200):
    S=np.linspace(0, 1, bins)
    expu=[]
    for s in S:
        u = U(futurewealth_logret(s, multiperiodret_random(rf, mean, sigma, k, n)[0], 
                                  multiperiodret_random(rf, mean, sigma, k, n)[1]), gamma)
        m = u.mean(axis=1).mean(axis=0)
        expu.append(m)
    sim = pd.DataFrame(np.column_stack([S, mexpu]), columns=['share', 'exputility'])
    #print('Stock allocation for Maximum Expected Utility: %2f.' %sim.loc[sim.exputility.argmax(), 'share'])
    return sim.loc[sim.exputility.argmax(), 'share']
```

```python
alloc_simplot(gw.Rfree.mean(), gw.CRSP_SPvw.mean(), gw.CRSP_SPvw.std(), 1, 200, 5)
```

#### constant expected returns

Extending to multile investment horizon does not change the allocation decision. 

```python
alloc = {}
for k in range(1, 6):
    alloc[k] = alloc_sim(gw.Rfree.mean(), gw.CRSP_SPvw.mean(), gw.CRSP_SPvw.std(), k, 200, 5)
alloc = pd.DataFrame(alloc, index=['alloc']).T
alloc.plot()
plt.show
```

This can be seen from the solution for optimal allocation:

$$\frac{E(r-r_{f})}{\gamma\sigma^2} = \frac{E(r^{e})}{\gamma\sigma^2} $$

with constent expected returns:
$$ \sigma^2_{k} = k\sigma^2 \\
E(r^{e}_{k}) = kE(r^{e}) $$
so the allocation doesn't change with investment horizon.

#### Variable and predictable expected returns

If returns are variable, for example, conditional expectation follows the mean reversion process above, asset allocation decision changes as investment horizon changes. Intuitively, as long horizon risk decrease, optimal allocation increases. 

##### Intertemporal CAPM
suppose we have $ V(W_{t+1}, z_{t+1})$. Value function depend on wealth and state variable for "investment opportunities", i.e., investors also care about future return $R_{j}, \quad j>1$.

$$V(W_{t},z_{t}) = max[u(c_{t})+\beta E_{t} V(W_{t+1}, z_{t+1})]$$
Take the first order derivative wrt $C_{t}$:
$$ u'(c_{t})-\beta E_{t}[ V_{W}(W_{t+1},z_{t+1})R^{W}_{t+1} ]=0$$
optimality also means that the effect of $w$ is the same on the left and right hand side of Bellman equation
$$ V_{W}(W_{t},z_{t})=\beta E_{t} [V_{W}(W_{t+1}, z_{t+1}) R^{W}_{t+1}] \quad (envelope theorem)$$

Therefore we have
$$ u'(c_{t})=V_{W}(W_{t},z_{t}) \quad \forall t $$
Consider the excess return formula from lecture 3
$$  E(R^{i})-R^{f}=-R^{f}cov(m,R^{i})=-\frac{cov(u'(c_{t+1}),R^{i}_{t+1})}{E[u'(c_{t+1})]}$$
it now becomes
$$  E(R^{i})-R^{f}= -\frac{cov( V_{W}(W_{t+1},z_{t+1}),R^{i}_{t+1})}{E[ V_{W}(W_{t+1},z_{t+1})]} $$



 Approximate $ V_{W}(W_{t+1}, z_{t+1})$ around $ V_{W}(W_{t}, z_{t})$,
 
$$ V_{W}(W_{t+1}, z_{t+1}) \approx  V_{W}(W_{t}, z_{t}) + V_{WW}(W_{t}, z_{t})\Delta W_{t+1}+V_{Wz}(W_{t}, z_{t})\Delta z_{t+1} $$
then excess return
$$ E_{t}(R_{i,t+1})-R_{t}^{f}=\gamma_{t} cov_{t}(R_{i,t+1},\Delta W_{t+1}/W_{t})+\lambda_{zt}cov_{t}(R_{i,t+1},\Delta z_{t+1}) $$ where
$\gamma_{t}=-\frac{WV_{WW}(W_{t},z_{t})}{V_{w}(W_{t},z_{t})}$ is the relative risk aversion. We can also write as
$$  E_{t}(R_{i,t+1}^{e})=\beta_{iM, t}(E_{t}R_{M,t+1}^{e})+\beta_{i\eta}(E_{t}R_{\eta,t+1}^{e}) $$
The second term is new, it represents the hedging motive (of future return opportunities) of the investors. $\eta$ here represents maximum hedging portfolio.

Merton's model indicates there is intertemporal hedging demand for investment opportunities. The market portfolio is a good hedge against its own reinvestment risk, long-horizon stock holding is larger than myopic demand.

#### Return Prediction with Vector Auto Regression (VAR)

In a system of predictive regression of $r_{t}$ and $dp_{t}$, if 
$$ r_{t} = \alpha_{0} + \beta_{11}r_{t-1} + \beta_{12} dp_{t-1} +\epsilon_{1t}$$
$$ dp_{t} = \alpha_{1} + \beta_{21}r_{t-1} + \beta_{22} dp_{t-1} + \epsilon_{2t}$$

we can then rewrite the system as 
$$
\begin{bmatrix}
    r_{t}       \\
    dp_{t}  \end{bmatrix} 
 = \begin{bmatrix}
   \alpha_{0}  \\
   \alpha_{1} \end{bmatrix}
 + 
 \begin{bmatrix}
   \beta_{11}  & \beta_{12}      \\
   \beta_{21}  & \beta_{22}  \end{bmatrix} 
\begin{bmatrix}
   r_{t-1}     \\
   dp_{t-1}  
\end{bmatrix} +
\begin{bmatrix}
    \epsilon_{1t}       \\
    \epsilon_{2t}      
\end{bmatrix}
$$

or $$ Y_{t} = A_{0} +A Y_{t-1} $$ where $Y$ is the stack of $r$ and $dp$ vectors.

We can impose structure on the VAR, suppose the model we want is:

$$ r_{t} = \alpha_{0} + \beta_{12} dp_{t-1} +\epsilon_{1t}$$
$$ dp_{t} = \alpha_{1} + + \beta_{22} dp_{t-1} + \epsilon_{2t}$$

we can then rewrite the system as 
$$
\begin{bmatrix}
    r_{t}       \\
    dp_{t}  \end{bmatrix} 
 = \begin{bmatrix}
   \alpha_{0}  \\
   \alpha_{1} \end{bmatrix}
 + 
 \begin{bmatrix}
  0  & \beta_{12}      \\
  0  & \beta_{22}  \end{bmatrix} 
\begin{bmatrix}
   r_{t-1}     \\
   dp_{t-1}  
\end{bmatrix} +
\begin{bmatrix}
    \epsilon_{1t}       \\
    \epsilon_{2t}      
\end{bmatrix}
$$

The above system can further be written as

$$
\begin{bmatrix}
    z'_{2}       \\
    \vdots\\
    z'_{T}      
\end{bmatrix} = \begin{bmatrix}
   1 & x'_{1}       \\
   \vdots & \vdots\\
    1 & x'_{T-1}      
\end{bmatrix} 
\begin{bmatrix}
    a'      \\
B'    
\end{bmatrix} +
\begin{bmatrix}
    \epsilon'_{2}       \\
    \vdots\\
    \epsilon'_{T}      
\end{bmatrix}
$$

where $$z_{t} =(r_{t}, x_{t})'$$
or simply $Z = XC+E$. 

$x_{t}$ is a vector of prediction variables. Here we just take $x_{t}=dp_{t}$. Z is the horizontal stack of $r$ and $dp$. In this form, coefficient C can be easily estimated as 

$$(X'X)^{-1}X'Z$$

```python
gw['lagdp'] = gw.dp.shift() 
data = gw.dropna()
data = data[['excessret', 'dp', 'lagdp']]
data['one'] =1
x = data[['one', 'lagdp']]
z = data[['excessret', 'dp']]
```

```python
C = np.linalg.inv(x.T.dot(x)).dot(x.T.dot(z))
C
```

```python
err = z.as_matrix() - (x.dot(C)).as_matrix()
SIGMA = err.T.dot(err)/(data.shape[0]-2) # degree of freedom (T-1)-n-1
SIGMA
```

the system of equations implied by these estimates are:

```python
print("r = {:.2f} + {:.2f}Lagdp".format(C[0,0], C[1,0]))
```

```python
print("dp = {:.2f} + {:.2f}Lagdp".format(C[0,1], C[1,1]))
```

we can then get predicted mean returns from these equations

```python
a1 = C[0,0]
b = C[1,0]
dp0 = data.tail(1)['dp'].values[0]
print("one month ahead predicted return %s"  % (a1 + b* dp0))

a2 = C[0,1]
fi = C[1,1]
dp1 = a2 + fi*dp0
print("two month ahead predicted return %s"  % (a1 + b*dp1))
```

In general, since $z_{t} = a +B x_{t-1}$, we can write 
$$z_{t} = a +B_{0} z_{t-1}$$ where
$$B_{0} = \begin{bmatrix} \begin{pmatrix} 0  \\  \vdots \\ 0 \end{pmatrix} & B \end{bmatrix}$$

Therefore 

$$ z_{T+1} = a + B_{0} z_{T } +\epsilon_{T+1} $$
$$ z_{T+2} = a + B_{0}a + B_{0}^{2} z_{T } +B_{0}\epsilon_{T+1} +\epsilon_{T+2} \\ 
\vdots \\
 z_{T+N} = a + B_{0}a + B_{0}^{2}a +\dots+B_{0}^{N-1}a + B_{0}^{N}z_{T} \\
 + \epsilon_{T+N} + B_{0}\epsilon_{T+N-1}+\dots +B_{0}^{N-1}\epsilon_{T+1} $$
 
 We can them calculate the conditional multiperiod mean return (sum) and variance. 

```python
B = np.asmatrix(np.array([[0, C[1,0]], [0, C[1,1]]]))
print("check 2 period prection:")
(B**2)[0,1]*dp0 + C[0,0] + B.dot(np.array([C[0,0], C[0,1]]))[0,0]
```

To see implication, let's ingore the constant term for a moment:

$$
\begin{align}
        r_{t+1} &=b\cdot dp_{t}+\varepsilon_{t+1}\\
        dp_{t+1}&=\phi \cdot dp_{t}+\delta_{t+1},
\end{align}
$$
The implied two period regression is
\begin{align*}r_{t+1}+r_{t+2}&= b\cdot dp_{t}+\varepsilon_{t+1}+b\cdot dp_{t+1}+\varepsilon_{t+2} \\
&=b\cdot dp_{t}+b(\phi \cdot dp_{t}+\delta_{t+1})+\varepsilon_{t+1}+\varepsilon_{t+2}\\
&= b(1+\phi)dp_{t}+(b\delta_{t+1}+\varepsilon_{t+1}+\varepsilon_{t+2}) \end{align*}
and \begin{equation*} r_{t+1}+r_{t+2}+r_{t+3}=b(1+\phi+\phi ^{2})dp_{t}+error\end{equation*}

* coefficients rise over horizon if $dp_{t}$ persistent ($\phi$ big). That is, small $b$ becomes big when horizon increases. 
* Multiperiod returns are correlated since $\delta_{t+1}$ and $\varepsilon_{t+1}$ are negatively correlated. This makes optimal allocation to increase with horizon. The opposing force is that estimation variance(risk) increase as horizon increases, which makes allocation smaller.  

Indeed, with annual data, b increases significantly.

### Market Timing

Another implication of return predictability is market timing: hold more stocks when expected returns are high, and fewer when expected returns are low (e.g. d/p, term spread, credit spread, oil price).

```python
Image('../img/strategic_allocation.JPG')
```

A simple strategy would be invest in risky assets only when predicted excess return is nonnegative, and invest in risk free asset otherwise. This requires an estimation of predictive regression first. 

```python
fit = smf.ols('excessret ~ TMS.shift() + DFY.shift() + dp.shift() ', data=gw).fit()
pret = pd.DataFrame(fit.predict(), index=gw.dropna().index, columns=['predict'])
pret = pret.join(gw)
pret['Mtiming'] = np.where(pret.predict>=0, pret.CRSP_SPvw, pret.Rfree)
```

```python
cumret= lambda x: (1+x).cumprod()-1
plt.rc('figure', figsize=(14, 10))
#fig, ax = plt.subplots()
pret[['Mtiming', 'CRSP_SPvw']].apply(cumret).plot()
plt.show()
```

It looks like the strategy performs well most of time. But, in fact, we were treating because we were using all the data throughout the periods. Our prediction is in-sample. Out of sample prediction is not easy. For details, see Campbell, John Y. and Samuel B. Thompson. 2008. Predicting excess stock returns out of sample: Can anything beat the historical average? The Review of Financial Studies 21(4): 1509- 1531.

```python
gwlag = pd.read_csv('../datalib/gw_shiller_m1926.csv')
gwlag['date'] = pd.to_datetime(gwlag.yyyymm.astype('str'), format='%Y%m') + pd.offsets.MonthEnd(0)
gwlag.drop('yyyymm', axis=1, inplace=True)
gwlag = gwlag.set_index('date')
gwlag['TMS'] = gwlag.lty - gwlag.tbl
gwlag['DFY'] = gwlag.AAA - gwlag.BAA
gwlag['dp'] = np.log(gwlag.D12/gwlag.pindex)
gwlag['ep'] = np.log(gwlag.E12/gwlag.pindex)
gwlag['excessret'] = gwlag.CRSP_SPvw - gwlag.Rfree
```

```python
# range of out of sample estmation
ran = pd.date_range('1953-01-31', '2011-12-30', freq='M')
osp = []
for i in range(len(ran)):
    # use data up to last period for estimation
    lp = ran[i] + pd.tseries.offsets.MonthEnd(-1)
    fit = smf.ols('excessret ~ TMS.shift() + DFY.shift() + ep.shift() + dp.shift()', data=gwlag[:lp]).fit()
    b = fit.params[1:].as_matrix()
    r = gwlag.loc[gwlag.index==ran[i], ['TMS', 'DFY', 'ep', 'dp']].values.reshape(4,)
    ospre = fit.params[0] + b.dot(r)
    if ospre>0:
        osp.append(np.asscalar(gwlag.loc[gwlag.index==ran[i], ['CRSP_SPvw']].values))
    else:
        osp.append(np.asscalar(gwlag.loc[gwlag.index==ran[i], ['Rfree']].values))
osp = pd.DataFrame(osp, index=ran, columns=['ospre'])
```

```python
pret2 = pret.join(osp, how='inner')
plt.rc('figure', figsize=(14, 10))
pret2[['ospre', 'CRSP_SPvw']].apply(cumret).plot()
plt.show()
```
