## Cross-section Prediction

### The question: 
whether there are stock characteristics (size, B/M, past returns, etc.) that allow us to predict different returns on different assets. 


The information set $I_{t}$ includes the cross-sectional characteristics of asset,  $C_{it}$, that is not risk loading $\beta_{i}$. If we use CAPM as benchmark model then,
$$ E_{t}R^{e}_{i,t+1}=\beta_{i}E_{t}(R^{e}_{M,t+1})+b_{i}C_{it}$$
Cross-sectional prediction tests whether $b_{i}=0$.

 $b_{i}\neq 0$ implies in
 
$$ R^{e}_{i,t}=\alpha_{i}+\beta_{i}R^{e}_{M,t}+\epsilon_{it}$$
$$ \alpha_{i}\neq 0 $$
That's just CAPM testing! In TS prediction regression of a panel, if we find $b/m_{it}$ predicts $R_{i,t+1}$, it has cross sectional prediction implications. TS and CS are the same thing, but they give us different *intuitions*.

### Size and B/M Sort

Many characteristics has been find to predict cross section returns. We'll investigate two characteristics: size and Book-to-Market. Let's get some portfolio return sorted on these two dimensions (note that size is proxied by market value of equity in Fama-French data library). 

 - B/M is measured in December of year $t-1$, 
 - size is measured in June of year $t$, and portfolios are formed annually in July of year $t$

```python
# data starts July 1926, we use post Jan 1963
start = datetime.datetime(1963, 1, 1)
end = datetime.datetime(2015, 12, 30)
df = web.DataReader("25_Portfolios_5x5", "famafrench", start, end)
df = pd.DataFrame(df[0])
df.index = df.index.to_timestamp('M')
```

```python
df.head()
```

```python
# import factors, let's concentrate on market factor first.
ds = web.DataReader("F-F_Research_Data_Factors", "famafrench", start, end)
ff = pd.DataFrame(ds[0])
ff.index = ff.index.to_timestamp('M')
ff.head()
```

```python
print('calculating excess returns')
# Here we use column operations
exret = df.apply(lambda x: x-ff['RF'])
exret.head()
```

#### Fitting the CAPM model 

```python
def ols_capm(Y, xcols=ff['Mkt-RF']):
    X = sm.add_constant(xcols)
    fit = sm.OLS(Y, X).fit()
    return fit.params[0], fit.params[1], fit.rsquared_adj #, fit.mse_total, fit.mse_resid
```

```python
capm_out = exret.apply(ols_capm)
alpha = capm_out.apply(lambda x: x[0])
beta = capm_out.apply(lambda x: x[1])
adjr2 = capm_out.apply(lambda x: x[2])
```

```python
def results_matrix(dfvar):
    # get index
    szindex = np.array([idx.split()[0] for idx in capm_out.index])
    szindex = szindex.reshape(5,5)[:,0].tolist()
    bmindex = np.array([idx.split()[1] for idx in capm_out.index])
    bmindex = bmindex.reshape(5,5)[0,:].tolist()
    results5by5 = pd.DataFrame(dfvar.values.reshape(5,5), index=szindex, columns=bmindex)
    return results5by5
```

```python
print('Portfolio Beta:')
betatb = results_matrix(beta)
betatb
```

```python
print('CAPM regression R-squared')
results_matrix(adjr2)
```

```python
print('CAPM alpha:')
results_matrix(alpha)
```

$\alpha_{i}=0$ is easily rejected. Is there any pattern in the estimation results?

#### Mean Excess Return vs CAPM Beta

```python
meanret = exret.mean()
meanrettb = results_matrix(meanret)
print('Mean Excess Return:')
meanrettb
```

- average (monthly) returns increase with B/M and decrease with size, something needs to be explained.
- recall that $\beta$ actually decrease with B/M and size
- High B/M stocks is usually called value stocks, low B/M stocks is growth stocks.

#### Fama-French solution
* Run time series regressions that include additional factors (portfolios of stocks) SMB,
HML, for each portfolio $i$
$$ R^{e}_{i,t}=\alpha_{i}+b_{i}R^{e}_{M,t}+s_{i}SMB_{t}+h_{i}HML_{t}+\epsilon_{it}$$
* Look across portfolios at the cross-sectional implication of this time-series regression (taking expectations)
$$ E(R^{e}_{i})=\alpha_{i}+b_{i}E(R^{e}_{M})+s_{i}E(SMB)+h_{i}E(HML)+\epsilon_{i}$$ to see if $\alpha_{i}\approx 0$.
* B/M is measured in December of year $t-1$, size is measured in June of year $t$, and portfolios are formed annually in July of year $t$

### FF factors
Solution of Fama-French
- First, 6 portfolios are formed based on the sorting of firm size (Small and Big) and B/M (High, Medium and Low) according to NYSE breaks:

|BMSize    | Small | Big |
|--------- |:-----:|:----:|
|High (30%) | SH   | BH |
|Medium(40%)| SM   | BM |
|Low (30%)  | SL   | BL |

- Second, for each portfolio, compute the monthly value-weighted return for the next 12 month.
- Mimicking portfolio factors:
  - $SMB=$ $\frac{r_{SH}+r_{SM}+r_{SL}}{3}-\frac{r_{BH}+r_{BM}+r_{BL}}{3}$
  - $HML=\frac{r_{BH}+r_{SH}}{2}-\frac{r_{BL}+r_{SL}}{2}$
  - Note that SMB and HML is already excess return
- 3 Factor Model: $R_{M}-R_{f}$, SMB, HML

#### 3 Factor Model
$R_{M}-R_{f}$, SMB, HML

```python
def ols_3f(Y, xcols=ff[['Mkt-RF', 'SMB', 'HML']]):
    X = sm.add_constant(xcols)
    fit = sm.OLS(Y, X).fit()
    return fit.params, fit.rsquared_adj #, fit.mse_total, fit.mse_resid
```

```python
ff3m = exret.apply(ols_3f)   # x[0] coefficients and x[1] rsquares
f3alpha = ff3m.apply(lambda x: x[0][0])
f3beta = ff3m.apply(lambda x: x[0][1])
f3smb = ff3m.apply(lambda x: x[0][2])
f3hml = ff3m.apply(lambda x: x[0][3])
ff3r2 = ff3m.apply(lambda x: x[1])
f3beta = results_matrix(f3beta)
f3smb = results_matrix(f3smb)
f3hml = results_matrix(f3hml)
```

```python
print('FF 3 factor model factor loadings')
pd.concat([f3beta, f3smb, f3hml], axis=1, keys=['b', 's', 'h'])
```

```python
print('R-Squared with Fama-Frend 3 Factor Model')
results_matrix(ff3r2)
```

- $b\approx 1$ after controlling for size and B/M
- Adjusted R-squares higher than the CAPM model 
  - Three-factor model explains almost 100% of return variation.

```python
def plotpred(modelpredictmean, actualmean, pltmodel, indextable):
    fig, ax = plt.subplots(figsize=(12, 12))
    for ptfl in indextable.columns:
        plt.plot(modelpredictmean[ptfl], actualmean[ptfl], '-o', label=ptfl)
    plt.title('Mean excess return v.s. Predicted Mean Excess Return with FF-3 Factor, %s' %pltmodel)
    plt.xlabel('Model Predicted Mean Excess Return')
    plt.ylabel('Actual Mean Excess Return')
    plt.xlim([0.2, 1.5])
    plt.ylim([0.2, 1.5])
    plt.legend()   
    plt.show()
```

###### Tautology?
- Is it a tautology to explain 25 B/M, size portfolios by 6 B/M, size portfolios?
- No. Works only if size, B/M are common factors. That is, stocks of similar sizes (B/M) comoves.
- example: Form portfolios by the stock's first letter, then form zero-investment portfolio by going long in A-M stocks and short
in N-Z stocks. Can this AMZ price other portfolios? No. Because with large N, idiosyncratic risk $\frac{1}{N}\sigma^{2}_{u}\rightarrow 0$, covariation emerges only if there
is common factor in returns.

The new paradigm:
$$ R_{i,t}-R_{f}=(\alpha_{i})+\beta_{i}(R_{M,t}-R_{f})+s_{i}SMB_{t}+h_{i}HML_{t}+\epsilon_{it} $$
Fama-French later (96, 08) find the 3 factor model can explain many other CAPM anomalies. This has implication on
 - Portfolio selection
 - Investment performance evaluation
 - Cost of capital estimation
 - Event studies

### Interpretations
What are the size and value factors?
 - rational explanations
  - HML proxies for aggregate "distress risk factor". Stocks become value after a string of bad news. Small size may proxy for "liquidity".
  - "state variables of concern" to investors, must be partly orthogonal to market return. Given market beta, they avoid stocks that go down in recessions, drive down prices and drive up expected returns. Now expected returns depend on tendency to go down in recessions as well as market beta.

 - APT Explanations
   - given the failure of CAPM to explain average returns, there would be arbitrage opportunities, if the returns on these mispriced stock did not have additional (size, B/M) factors in their CAPM residuals.

### Return Reversal and Momentum

#### Return Reversal
De Bondt and Thaler (1995, JF) form portfolios on the basis of past 3 to 5 year returns. They show that long term losers outperform long term winners in the next 3 to 5 years.

#### Return Momentum
Jegadeesh and Titman (1993) show that trading strategies based on short term losers and winners generate superior returns.
 - Each month, buy stocks that are winners in the past 3 to 12 months, and hold them for the next 3 to 12 months. Symmetrically, sell the short-term losers.
 - On average, the abnormal return is about 1% monthly. Persisted in later sample.

 - Return reversal can be related to B/M anomaly. A string of bad news gives a low price and a high B/M. Hence, given the value premium, a stock with low past returns (value) should earn high expected returns.
 - Momentum cannot be explained by FF 3 factor model. No good risk-based explanation.
