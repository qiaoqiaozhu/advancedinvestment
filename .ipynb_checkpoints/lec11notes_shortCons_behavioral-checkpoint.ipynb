## Heterogeneous Beliefs and short sale constraint

If agents have heterogeneous Beliefs and short sale is not allowed, then stock price cannot incorporate all information, and will deviate from “fundamental” value even if agents are all “rational”. Let build a simple model to show that.

### Harrison and Kreps (QJE 1978) Model
A simple asset with two state ($d_{t}$),  $d_{t}=0 $ or $1$. Constant discount factor $\gamma$, under which agents are risk neutral. Let $q(d,d')$ be transition probability $Pr(d_{t+1} = d' | d_{t} = d)$ Transition matrix is then
$$
Q=\left[
\begin{array}{cc}
q(0,0) & q(0,1) \\
q(1,0) & q(1,1)
\end{array}%
\right]
$$
where Rows denote current state, columns denote future state.

Present Values

$$
E_{t}d_{t+1}=0*Pr(d_{t+1} = 0 | d_{t} )+1*Pr(d_{t+1} = 1| d_{t} )=Q\left[
\begin{array}{c}
0\\
1
\end{array}%
\right]
$$

$$
E_{t}d_{t+i}=Q^{i}\left[
\begin{array}{c}
0\\
1
\end{array}
\right]
$$

$$
E_{t}\sum_{i=1}^{\infty}d_{t+i}=(\gamma Q+\gamma^{2}Q^{2}+\dots)\left[
\begin{array}{c}
0\\
1
\end{array}
\right]
= \gamma Q(I-\gamma Q)^{-1}\left[
\begin{array}{c}
0\\
1
\end{array}
\right]
$$


Heterogeneous Beliefs


* Two investors, a & b, with the same $\gamma=0.75$ but different Q:

$$
 Q^{a}=\left[
\begin{array}{cc}
1/2 & 1/2 \\
2/3 & 1/3
\end{array}
\right]
\quad Q^{b}=\left[
\begin{array}{cc}
2/3 & 1/3 \\
1/4 & 3/4
\end{array}
\right]
$$

* Investor a is more optimistic in initial state 0, b is more optimistic in initial state 1
* their perceptions of the fundamental value:
$$
V^{a}
= \gamma Q^{a}(I-\gamma Q^{a})^{-1}\left[
\begin{array}{c}
0\\
1
\end{array}
\right]
=\left[\begin{array}{c}
1.33\\
1.22
\end{array}
\right]
$$

$$
V^{b}
= \gamma Q^{b}(I-\gamma Q^{b})^{-1}\left[
\begin{array}{c}
0\\
1
\end{array}
\right]
=\left[\begin{array}{c}
1.45\\
1.91
\end{array}
\right]
$$


```python
Qa = np.array([[1/2,1/2], [2/3, 1/3]])
Qb = np.array([[2/3,1/3], [1/4, 3/4]])
d = np.array([[0],[1]])
I = np.identity(2)
Va = 0.75* Qa.dot(np.linalg.inv((I-0.75*Qa))).dot(d)
Vb = 0.75* Qb.dot(np.linalg.inv((I-0.75*Qb))).dot(d)
print('Va = ', Va)
print('Vb = ', Vb)
```

#### Equilibrium Prices

* Does this mean that  investor b always hold the stock, and that prices equal investor b's assessment of fundamental value?


* No: with prices 1.45 and 1.91, investor a can buy in state 0 and sell in state 1. His expected payoff

$$ (0.75*1/2+0.75^{2}*(1/2)^{2}+\dots)(1+1.91)=1.75>1.45$$
And if price rises to 1.75 in state 0, investor b will buy in state 1 and sell when state 0 occurs.

$$\gamma(\frac{1}{4}1.75+\frac{3}{4}1)+\gamma^{2}(\frac{3}{4}\frac{1}{4}1.75+\frac{3}{4}\frac{3}{4}1)+\dots=2.03>1.91 $$

```python
init = 0.5*0.75*(1+1.91)
mul = 1/(1-0.75*0.5)
print(init*mul)
init = (1/4*1.75 + 3/4)*0.75
mul = 1/(1-0.75**2)
print(init*mul)
```

A consistent price scheme is such that
$$ p(d)=\max_{s=a, b}E^{s}(\gamma d_{t+1}+\gamma p(d_{t+1})|d_{t}) $$

In this example
$$ p(0)=0.75 \max (\frac{1}{2}p(0)+\frac{1}{2}(1+ p(1)), \frac{2}{3}p(0)+\frac{1}{3}(1+ p(1))) $$

$$ p(1)=0.75 \max (\frac{2}{3}p(0)+\frac{1}{3}(1+ p(1)), \frac{1}{4}p(0)+\frac{3}{4}(1+ p(1))) $$
$\rightarrow p(0)=1.846, p(1)=2.077$  

* Price exceeds fundamental as perceived by either investor.

```python
# p1:x p0:y
x = np.linspace(0, 3)
ap0 = 0.75*(0.5*(1+x))/(1-0.75*0.5)
bp0 = 0.75*((1/3)*(1+x))/(1-0.75*(2/3))
ap1 = ((1-0.75*(1/3))*x -0.75*(1/3))/(0.75*(2/3))
bp1 = ((1-0.75*(3/4))*x -0.75*(3/4))/(0.75*(1/4))
fig, ax = plt.subplots()
plt.plot(x, ap0, label = 'Person a, P(d=0)')
plt.plot(x, bp0, label = 'Person b, p(d=0)')
plt.plot(x, ap1, label = 'Person a, p(d=1)')
plt.plot(x, bp1, label = 'Person b, p(d=1)')
plt.plot(2.077, 1.846, marker='o', markersize=8, color="red")
plt.title('Heterogenous Belief')
plt.legend()
plt.xlabel('p(1)')
plt.ylabel('p(0)')
plt.show()
```

Crutial in this example, there is short sale constraint, agents can only try to profit through purchasing stocks. Short sale makes overvaluation possible. 

### Investor Sentiment


Investor sentiment has larger effects on securities whose valuations are highly subjective and difficult to arbitrage (Baker & Wugler (2006)), this bias is difficult to correct with short constraint. 

When beginning-of-period proxies for sentiment are low, subsequent re- turns are relatively high for small stocks, young stocks, high volatility stocks, un-profitable stocks, non-dividend-paying stocks, extreme growth stocks, and distressed stocks. When sentiment is high, on the other hand, these categories of stock earn relatively low subsequent returns.

```python
sent = pd.read_csv('../datalib/bw_sent.csv')
sent['date'] = pd.to_datetime(sent.date.astype('str'), format="%Y%m")
sent = sent.set_index('date')
sent.index = sent.index.to_period('M').to_timestamp('M')
plt.figure(figsize=(14,10))
plt.plot(sent.sent)
plt.axhline(sent.sent.mean(), color='r', linestyle='--')
plt.xlabel('T')
plt.ylabel('Sentiment Index')
plt.title('Sentiment Index')
plt.legend()
plt.show()
```

```python
def sasdate(dvar):
    return pd.to_timedelta(dvar, unit='D') + pd.Timestamp('1960-1-1')
sp500 = pd.read_sas('../datalib/sp500.sas7bdat')
sp500['Date'] = sasdate(sp500.caldt) + pd.offsets.MonthEnd(0)
sp500 = sp500[sp500.Date>'1963-1-1']
sp500 = sp500[['Date', 'VWRETD', 'EWRETD', 'SPRTRN', 'SPINDX']].set_index('Date')
# log moving average detrend
logts = np.log(sp500.SPINDX)
moving_avg = logts.rolling(window=36,center=False).mean()
diff = 8*(logts - moving_avg).dropna()
plt.figure()
diff.plot(label="log moving average detrend S&P 500")
sent.sent.plot(label='Sentiment Index')
plt.legend()
plt.show()
```

Stambaugh et al (2012): PCA of anomalies is negatively correlated with lag sentiment, this is largely driven by the short leg.

```python
anomaly = pd.read_excel('../datalib/longshortreturns.xlsx', sheetname='Sheet1', na_values='NaN')
anomaly['day'] =1
anomaly['Date'] = pd.to_datetime(anomaly[['year', 'month', 'day']])
anomaly['Date'] = anomaly['Date'] + pd.offsets.MonthEnd(0) 
anomaly = anomaly.drop(['day', 'year', 'month'], axis=1)
anomaly = anomaly.set_index('Date')
colv = anomaly.columns.tolist()
namestem = []
for col in colv:
    stem = col.split('_')[-1]
    if stem not in namestem:
        namestem.append(stem)
namestem.remove('ave')
premium = pd.DataFrame()
for col in namestem:
    rescol = 'ret_' + col
    longcol = 'ret_long_' + col
    shortcol = 'ret_short_' + col
    premium[rescol] = anomaly[longcol] - anomaly[shortcol]

premium.head(3)
```

```python
from sklearn.preprocessing import scale, StandardScaler
from sklearn.decomposition import PCA as sklearnPCA
X = premium.dropna()
X_std = StandardScaler().fit_transform(X)
sklearn_pca = sklearnPCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(X_std) 
lspcadf = pd.DataFrame(-Y_sklearn, columns=['pca1', 'pca2'], index=X.index).join(sent, how='inner')

lspcadf.head(3)
```

```python
results = smf.ols('pca1 ~ sent.shift()', data = lspcadf).fit(cov_type='HC0')
print(results.summary())
```

```python
short = anomaly.filter(regex=('short_(?!ave)')).dropna()
shortskl = sklearnPCA(n_components=2)
spca = shortskl.fit_transform(StandardScaler().fit_transform(short)) 
spcadf = pd.DataFrame(-spca, columns=['pca1', 'pca2'], index=short.index).join(sent)
results = smf.ols('pca1 ~ sent.shift()', data = spcadf).fit(cov_type='HC0')
print(results.summary())
```

### Behavioral Biases

[List of cognitive biases](https://en.wikipedia.org/wiki/List_of_cognitive_biases)

#### Seeing things that are not there

Which of the following sequences is more likely to occur
when a fair coin is tossed - HHHTTT or HTHTTH? 



* Many people erroneously believe that the second sequence is more likely than the first. Since the first one appears systematic. In fact, the two sequences in this question are equally likely to occur.
* Many people will be far too quick to perceive causal regularity in random sequences of events. Tend to see patterns in historical data that are not really there.
* "Hot hand" fallacy


#### Similarity vs Probability
* Decision makers use similarity or representativeness as a proxy for probabilistic thinking.
* But, sometimes similarity poorly predicts true probability.


Linda is 31 years old, single, outspoken, and very bright.
She majored in philosophy. As a student, she was deeply
concerned with issues of discrimination and social justice,
and also participated in anti-nuclear demonstrations. Rank the following probability?
 1. Linda is a bank teller.
 2. Linda is a psychiatric social worker.
 3. Linda is a bank teller and is active in the feminist movement.

* 80\%-90\% persons rank (3) higher than (1).
* But, $P(feminist \cap bankteller) \leq P(bankteller)$. Similarity relations do not follow the conjunction rule.
*  similar biases leads to insensitivity to base prob. and to small sample size


### Prospect Theory

Prospect Theory (Kahneman and Tversky 1979): people care about changes not levels and use a weighted value function to think about risky decisions

$$  U=\sum_{i=1}^{S} \pi(p)v(x)$$
where $\pi(p)$ is subjective weight on an event with objective probability $p$. $v(x)$ is the value of a change of magnitude $x$.

* a gamble $(60\% +\$10, 40\% -\$10)$ would have prospect theoretic value: $\pi(0.6)v(10)+\pi(0.4)v(-10)$, contrast to expected utility $0.6u(w+10)+0.4u(w-10) $

* $v(x)$, $x$ relative to status quo, framing matters
* $v(x)$ has a kink around 0. A loss of size $\Delta$ is (e.g. 2x) as important as a gain of size $\Delta$

##### Loss Aversion
 People are indifferent between
*  status quo
* 2/3 chance of +10 and 1/3 chance of -10
 or
* status quo
* 1/2 chance of +20 and 1/2 chance of -10

Small risk now matters, loss aversion discourages trade.

#### Evidence on Loss Aversion

* People are predisposed to "getting even". Also known as disposition effect: Propensity to lock in sure gains
than to lock in sure loss.
* Individuals are more likely to sell stocks that have risen in price
rather than those that have fallen in price (Odean 2001).
* People do not treat gains and losses symmetrically.

### Smart money and limit to arbitrage
*  Investors often display behavioral biases
* But asset price is determined by marginal investors.
* As long as there are some sophisticated investors, therefore marginal investors are ''smart money'', arbitrage opportunities should not persist.
* Limited arbitrage is necessary for behavioral finance to affect equilibrium asset pricing.
