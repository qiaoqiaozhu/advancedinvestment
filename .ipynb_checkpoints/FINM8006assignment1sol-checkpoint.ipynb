* Q1

suppose that your desired rate of return is 3% in the month, and you want to acheive mean-variance efficiency using **just the 5 stocks**. Find the **optimal weights on the stocks, and the expected mean and standard deviation of your portfolio.** 

```python
## function to find mean-variance without rf
def meanvar(mu, rbar=rbar0, cov=cov0):
    ones = np.ones(rbar.shape)
    A = rbar.T.dot(inv(cov)).dot(rbar)
    B = rbar.T.dot(inv(cov)).dot(ones)
    C = ones.T.dot(inv(cov)).dot(ones)
    omega = (C*mu-B)/(A*C-B**2)*dot(inv(cov), rbar) + (A-B*mu)/(A*C-B**2)*dot(inv(cov), ones)
    sigmahut = np.sqrt(omega.transpose().dot(cov).dot(omega))
    rp = omega.T.dot(rbar)
    return omega, rp, sigmahut
```

```python
alloc1 = meanvar(0.03)
```

```python
w1 = DataFrame(alloc1[0], index=stocks.PERMNO.as_matrix(), columns=['weight'])
w1 = w1.join(stocks.set_index('PERMNO'))
print('weights\n')
w1
```

### Mean-Variance With Risk Free Assets

* Q4

You've seen the derivation of mean-variance optimal allocation with risk free assets in class notes. Derive an expression of **relative weights in risky assets (out of total investment in risky assets)** for a portfolios of N risky assets. Use the same notation as the lecture notes. 

$$ \min_{\omega} \frac{1}{2}\omega' \Sigma \omega \qquad
\text{s.t. }  \omega'(\bar{R}-R_{f}\iota)=\mu-R_{f} $$

Solving FOC we get $$\omega =\lambda \Sigma^{-1}(\bar{R}-R_{f}\iota) $$
$$\lambda =\frac{\mu-R_{f}}{(\bar{R}-R_{f}\iota)'\Sigma^{-1}(\bar{R}-R_{f}\iota)} $$

$$ w = \frac{\omega_{i}}{\iota'\omega'} = \frac{\lambda \Sigma^{-1}(\bar{R}-R_{f}\iota)}{\lambda \iota' \Sigma^{-1}(\bar{R}-R_{f}\iota)}= \frac{ \Sigma^{-1}(\bar{R}-R_{f}\iota)}{\iota' \Sigma^{-1}(\bar{R}-R_{f}\iota)}$$

Now that you can invest (or borrow) from risk free rate. 

* Q5

Use the math expression you derive above to find the optimal relative weights in the five stocks in January 1997.  


```python
rf0 = msf.loc[investmonthEnd, 'rf'].mean()
rf0
```

define function for solution given the expressions above:

```python
def meanvar_rfra(rf, rbar=rbar0, cov=cov0):
    ones = np.ones(rbar.shape)
    w = inv(cov).dot(rbar-rf*ones)/ones.T.dot(inv(cov)).dot(rbar-rf*ones)
    return w
```

```python
mvalloc = meanvar_rfra(rf0, rbar=rbar0, cov=cov0)
w4 = DataFrame(mvalloc, index=stocks.PERMNO.as_matrix(), columns=['weight'])
w4 = w4.join(stocks.set_index('PERMNO'))
print('weights\n')
w4
```

```python
def findparam(year, month, day):
    investmonth = pd.datetime(year, month, day)
    investmonthEnd = investmonth + MonthEnd(0)
    mrange = pd.date_range(end=investmonth, periods=36, freq='M')
    mrangeret = msf.loc[msf.index.isin(mrange), ['PERMNO', 'RET']]
    retbystock = mrangeret.groupby('PERMNO')['RET']
    retmat = mrangeret.reset_index().set_index(['PERMNO', 'date']).unstack('PERMNO')
    rbar = retbystock.mean().as_matrix()
    cov = retmat.cov().as_matrix()
    rf = msf.loc[investmonthEnd, 'rf'].mean()
    return rbar, cov, rf
```

* Q7

Evaluate the success of your mean-variance strategy in the two periods. Can you explain why your strategy breaks down in the second period?

```python
## your loop month over the index can be written as 
# freq='MS' means month start
pd.date_range(start='2004-1-1', end='2007-7-1', freq='MS')
```

The key here is to keep track of rw, return to your wealth we find out before, essentially putting the code of rw before in a loop

```python
def meanvar_rfloop(loopstart, loopend):
    loopm = pd.date_range(start=loopstart, end=loopend, freq='MS')
    alpha = 0.8
    wstar = pd.DataFrame(columns=['w1','w2','w3','w4','w5' ])
    pret = []
    for investmonth in loopm:
        mrange = pd.date_range(end=investmonth, periods=36, freq='M')
        mrangeret = msf.loc[msf.index.isin(mrange), ['PERMNO', 'RET']]
        rbar =mrangeret.groupby('PERMNO')['RET'].mean().as_matrix()
        cov = mrangeret.reset_index().set_index(['PERMNO', 'date']).unstack('PERMNO').cov().as_matrix()
        investmonthEnd = investmonth+MonthEnd(0)
        rf = msf.loc[investmonthEnd, 'rf'].mean()
        ones = np.ones(rbar.shape)
        w = inv(cov).dot(rbar-rf*ones)/ones.T.dot(inv(cov)).dot(rbar-rf*ones)
        W = DataFrame(w, index=['w1','w2','w3','w4','w5' ]).T
        cmret = msf.loc[investmonthEnd, 'RET'].as_matrix()
        mretp = w.T.dot(cmret)
        wstar = wstar.append(W)
        rw = (1-alpha)*rf + alpha * mretp
        pret.append(rw)
    wstar.index = loopm
    return DataFrame(pret,index=loopm, columns=['pret']), wstar
```

```python
def oneovern_loop(loopstart, loopend):
    loopm = pd.date_range(start=loopstart, end=loopend, freq='MS')
    alpha = 0.8
    pret = []
    for investmonth in loopm:
        investmonthEnd = investmonth+MonthEnd(0)
        rf = msf.loc[investmonthEnd, 'rf'].mean()
        cmret = msf.loc[investmonthEnd, 'RET'].mean()
        rw = (1-alpha)*rf + alpha * cmret
        pret.append(rw)
    wstar.index = loopm
    return DataFrame(pret,index=loopm, columns=['ewret'])
```

```python
pret, wstar = meanvar_rfloop('2004-1-1', '2007-7-1')
```

```python
poneovern = oneovern_loop('2004-1-1', '2007-7-1')
```

```python
cumuret = pret.apply(lambda x: (1+x).cumprod())
cumueqret = poneovern.apply(lambda x: (1+x).cumprod())
plt.figure()
pd.concat([cumuret, cumueqret], axis=1).plot()
plt.show()
```

```python
pret, wstar = meanvar_rfloop('2007-1-1', '2010-7-1')
poneovern = oneovern_loop('2007-1-1', '2010-7-1')
cumuret = pret.apply(lambda x: (1+x).cumprod())
cumueqret = poneovern.apply(lambda x: (1+x).cumprod())
plt.figure()
pd.concat([cumuret, cumueqret], axis=1).plot()
plt.show()
```

Consider an investor with risk aversion $\gamma$
whose preferences are fully described by the mean and variance of a chosen portfolio such that his expected utility is characterized by:
\begin{align*}
E(U)= E(R_{p})-\frac{\gamma}{2}VAR(R_{p})
\end{align*}
where $R_{p}$ is the return of her investment portfolio.


Suppose there are N risky assets with expected return $\bar R$ and return variance-covariance of $\Sigma$. $\bar R$ is $N \times 1$ vestor and $\Sigma$ is a $N \times N$ matrix. Risk-free asset is available with return $R_{f}$. Let $\omega$ be a $N \times 1$ vector of portfolio weights (as a reportion of wealth) on risky assets. 

* Q8

Solve for the optimal  $\omega$ for this investor by maximizing expected utility. Show that you can get the same solution in terms of relative weight of risky assets in the risky portfolio as the mean-variance with risk free rate above. Discuss how risk aversion is related to the weight on risky assets based on your solution. 

\begin{align*}
\max_{\omega} \omega'\bar R+(1-\omega'\iota)R_{f}-\frac{\gamma}{2}\omega' \Sigma \omega
\end{align*}
FOC: \begin{equation*}\bar R-R_{f}\iota-\gamma \Sigma \omega=0 \end{equation*}
Solving the FOC, we get \begin{equation*}\omega=\frac{1}{\gamma}\Sigma^{-1} (\bar R-R_{f}\iota)\end{equation*}

The higher the $\gamma$, the lower the $\omega$. Notice that $\omega_{i}$ proportionally change with $\gamma$. 

The relative weight is 
$$ w= \frac{\Sigma^{-1} (\bar R-R_{f}\iota)}{\iota'\Sigma^{-1} (\bar R-R_{f}\iota)} $$
