```python
from __future__ import division
from __future__ import print_function
import numpy as np
from pandas import Series, DataFrame
from pandas.tseries.offsets import MonthEnd
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import datetime
plt.rc('figure', figsize=(16, 10))
import re
from scipy.stats import ttest_1samp
```

```python
#fin = pd.read_csv('/Users/qiaoqiaozhu/Documents/Data/csmar_financials.csv')
fin = pd.read_csv('csmar_financials.csv')
fin.columns = ['STKCD', 'date', 'TA', 'TL', "BE"]
fin.date = pd.to_datetime(fin.date.astype('str'), format='%Y%m%d') + MonthEnd(0)
fin = fin.loc[fin.BE > 0] # require positive book value
fin = fin.sort_values(["STKCD", 'date', 'BE'])
fin = fin.groupby(['STKCD', 'date']).last().reset_index()
fin.head(3)
```

```python
#ret = pd.read_csv('/Users/qiaoqiaozhu/Documents/Data/csmar2000to2016febencode.csv', encoding='utf-8')
ret = pd.read_csv('csmar2000to2016febencode.csv', encoding='utf-8')

ret.columns = ['STKCD', 'MARKETTYPE', 'TRDMNT', 'PriceClose', 'tradableME', 'ME', 'ret', 'retxd']
ret['ME'] = ret['ME']*1000    # Msmvttl, monthly ME is in CNY1000 units
ret.TRDMNT = pd.to_datetime(ret.TRDMNT.astype('str'), format='%Y%m') + MonthEnd(0)
# markettype 1 Shanghai A, 4, Shengzhen A
ret = ret.loc[ret.PriceClose>=2]  # require price to be higher than 2
ret = ret.loc[ret.ME>0.5e+09]   # require minimum size to be 500 mill CNY
ret.head(3)
```

```python
# year is the portfolio formation year
juneme = ret.loc[ret.TRDMNT.dt.month==6, ['STKCD', 'MARKETTYPE', 'TRDMNT', 'ME']]
juneme['year'] = juneme.TRDMNT.dt.year
juneme = juneme.drop('TRDMNT', axis=1)
juneme.head()
```

```python
decbe = fin.loc[fin.date.dt.month==12, ['STKCD', 'date', 'BE']]
decbe['year'] = fin.date.dt.year + 1
decbe = decbe.drop('date', axis=1)
decbe.head()
```

```python
# SSE ME and BTM for cutoff 
mebe = juneme.merge(decbe, on=['STKCD', 'year'], how='inner')
mebe['BTM'] = mebe.BE / mebe.ME
mebe = mebe.loc[mebe.MARKETTYPE==1]  #only shanghai market for cutoff calculation
mebe.head(3)
```

```python
mebe[['ME', 'BE', 'BTM']].describe()
```

```python
ret['year'] = np.where(ret.TRDMNT.dt.month>=7, ret.TRDMNT.dt.year, ret.TRDMNT.dt.year-1)
```

```python
stkret = ret.loc[ret.TRDMNT>='2000-7-30', ['STKCD', 'MARKETTYPE', 'TRDMNT', 'year', 'ret']]
stkret = stkret.merge(juneme[['STKCD', 'year', 'ME']], on=['STKCD', 'year'], how='inner')
stkret = stkret.merge(decbe[['STKCD', 'year', 'BE']], on=['STKCD', 'year'], how='inner')
stkret['BTM'] = stkret.BE/stkret.ME
stkret.head(3)
```

```python
btmcutoff = mebe.groupby('year')['BTM'].quantile([ .3, .7,]).unstack()
btmcutoff.columns = ['highcut', 'lowcut']
btmcutoff = btmcutoff.reset_index()
sizecutoff = pd.DataFrame(mebe.groupby('year')['ME'].quantile(0.5))
sizecutoff.columns = ['largecut']
sizecutoff = sizecutoff.reset_index()
```

```python
stkret = stkret.merge(btmcutoff, on=['year'], how='inner')
stkret = stkret.merge(sizecutoff, on=['year'], how='inner')
```

```python
stkret.loc[stkret.BTM>=stkret.highcut, 'bmport'] = 'high'
stkret.loc[(stkret.lowcut<=stkret.BTM) & (stkret.BTM<stkret.highcut), 'bmport'] = 'medium'
stkret.loc[stkret.BTM<stkret.lowcut, 'bmport'] = 'low'
```

```python
stkret.loc[stkret.ME>=stkret.largecut, 'szport'] = 'big'
stkret.loc[stkret.ME<stkret.largecut, 'szport'] = 'small'
```

```python
stkret['bmsz'] = stkret.bmport + stkret.szport
stkret.head(3)
```

```python
g = stkret.groupby(['year', 'bmsz'])
stkret['w'] = stkret.ME / g.ME.transform("sum")
stkret['wret'] = stkret.ret * stkret.w
```

```python
vwportret = stkret.groupby(['TRDMNT', 'bmsz'])['wret'].sum().reset_index()
```

```python
h = vwportret.loc[vwportret.bmsz.str.contains('high'), ['wret', 'TRDMNT']].groupby('TRDMNT')['wret'].mean()
l = vwportret.loc[vwportret.bmsz.str.contains('low'), ['wret', 'TRDMNT']].groupby('TRDMNT')['wret'].mean()
hml = h-l
hml.name = 'HML'
```

```python
s = vwportret.loc[vwportret.bmsz.str.contains('small'), ['wret', 'TRDMNT']].groupby('TRDMNT')['wret'].mean()
b = vwportret.loc[vwportret.bmsz.str.contains('big'), ['wret', 'TRDMNT']].groupby('TRDMNT')['wret'].mean()
smb = s-b
smb.name = 'SMB'
```

```python
assignbm = lambda x: pd.qcut(x, 5, labels=['bm1', 'bm2', 'bm3', 'bm4', 'bm5'], retbins=False)
stkret['assignbm'] = stkret.groupby('TRDMNT')['BTM'].apply(assignbm)
assignsz = lambda x: pd.qcut(x, 5, labels=['sz1', 'sz2', 'sz3', 'sz4', 'sz5'], retbins=False)
stkret['assignsz'] = stkret.groupby('TRDMNT')['ME'].apply(assignsz)
stkret['port'] = stkret.assignbm.astype('str') + stkret.assignsz.astype('str')
stkret = stkret[['STKCD', 'MARKETTYPE', 'TRDMNT', 'ret', 'port']]
stkret.head(3)
```

```python
bmsz = stkret.groupby(['TRDMNT', 'port'])['ret'].mean()
bmsz = bmsz.reset_index().set_index('TRDMNT')
bmsz = bmsz.join(hml)
bmsz = bmsz.join(smb)
bmsz.head(3)
```

```python
#mktret = pd.read_csv('/Users/qiaoqiaozhu/Documents/Data/csmarmktret.csv')
mktret = pd.read_csv('csmarmktret.csv')
mktret = mktret[mktret.MARKETTYPE==1]
mktret.TRDMNT = pd.to_datetime(mktret.TRDMNT.astype('str'), format='%Y%m') + MonthEnd(0)
mktret = mktret[mktret.TRDMNT>='2000-7-31']
mktret = mktret.drop('MARKETTYPE', axis=1)
mktret.rename(columns={'MRETWDTL': 'mktret', 'MRETMDTL':'mktretxd'}, inplace=True)
mktret = mktret.set_index('TRDMNT')
mktret.head(3)
```

```python
portret = bmsz[['port',  'ret']].reset_index().set_index(['port', 'TRDMNT']).unstack('port')
portret.columns = portret.columns.droplevel()
portret.head(3)
```

```python
def ols_capm(Y, xcols=mktret['mktret']):
    X = sm.add_constant(xcols)
    fit = sm.OLS(Y, X).fit()
    return fit.params[0], fit.params[1]
```

```python
capm_out = portret.apply(ols_capm)
alpha = capm_out.apply(lambda x: x[0])
beta = capm_out.apply(lambda x: x[1])
```

```python
def results_matrix(dfvar):
    # get index
    szindex = np.array([re.findall(r'(?<=\d)[a-z]{2}\d', idx)[0] for idx in capm_out.index]) 
    szindex = szindex.reshape(5,5)[0].tolist()
    bmindex = np.array([idx.split('s')[0] for idx in capm_out.index])
    bmindex = bmindex.reshape(5,5)[:,0].tolist()
    results5by5 = pd.DataFrame(dfvar.values.reshape(5,5), index=bmindex, columns=szindex)
    return results5by5
```

```python
print('Portfolio Beta:')
betatb = results_matrix(beta)
betatb
```

```python
print('CAPM alpha:')
results_matrix(alpha)
```

```python
meanrettb = results_matrix(portret.mean())
print('Mean Excess Return:')
meanrettb
```

```python
def momentum(stkret, j=12, k=6, stkcd='STKCD', tradedate='TRDMNT', ret='ret'):
    origret = stkret[[stkcd, tradedate, ret]]
    momstk = origret.set_index(tradedate)
    momstk = momstk.groupby(stkcd)[ret].rolling(window = j).apply(lambda x: np.prod(1+x)-1).reset_index().dropna()
    momstk = momstk.groupby(tradedate).filter(lambda x: len(x)>10)  #filter dates less than 10 obs
    assignmom = lambda x: pd.qcut(x, 10, labels=['dec'+str(i) for i in range(1,11)], retbins=False)
    momstk['dec'] = momstk.groupby(tradedate)[ret].apply(assignmom)
    momstk = momstk.loc[momstk.dec.isin(['dec1', 'dec10']), [stkcd, tradedate, 'dec']]
    holdport = {}
    for m in range(k):
        holdport[m] = momstk.copy()
        holdport[m][tradedate] = holdport[m][tradedate] + pd.DateOffset(months=m) + MonthEnd(0)
        holdport[m] = holdport[m].merge(origret, on=[stkcd, tradedate], how='inner')
        holdport[m] = holdport[m].groupby(['dec', tradedate])[ret].mean()
    holdmean = pd.concat([holdport[i] for i in range(k)], axis=1).mean(axis=1)
    dec10 = holdmean.loc[holdmean.index.get_level_values('dec') == 'dec10']
    dec10.index = dec10.index.droplevel(level=0)
    dec10.name = 'long'
    dec1 = holdmean.loc[holdmean.index.get_level_values('dec') == 'dec1']
    dec1.index = dec1.index.droplevel(level=0)
    dec1.name = 'short'
    momret = pd.concat([dec10, dec1], axis=1)
    momret['mom'] = momret.long - momret.short
    return momret.mom.mean(), ttest_1samp(momret.mom, 0)[0]
```

```python
def umd(stkret, j=12, k=1, stkcd='STKCD', tradedate='TRDMNT', ret='ret'):
    origret = stkret[[stkcd, tradedate, ret]]
    momstk = origret.set_index(tradedate)
    momstk = momstk.groupby(stkcd)[ret].rolling(window = 12).apply(lambda x: np.prod(1+x)-1).reset_index().dropna()
    momstk = momstk.groupby(tradedate).filter(lambda x: len(x)>10)  #filter dates less than 10 obs
    assignmom = lambda x: pd.qcut(x, 10, labels=['dec'+str(i) for i in range(1,11)], retbins=False)
    momstk['dec'] = momstk.groupby(tradedate)[ret].apply(assignmom)
    momstk = momstk.loc[momstk.dec.isin(['dec1', 'dec10']), [stkcd, tradedate, 'dec']]
    holdport = {}
    for m in range(k):
        holdport[m] = momstk.copy()
        holdport[m][tradedate] = holdport[m][tradedate] + pd.DateOffset(months=m) + MonthEnd(0)
        holdport[m] = holdport[m].merge(origret, on=[stkcd, tradedate], how='inner')
        holdport[m] = holdport[m].groupby(['dec', tradedate])[ret].mean()
    holdmean = pd.concat([holdport[i] for i in range(k)], axis=1).mean(axis=1)
    dec10 = holdmean.loc[holdmean.index.get_level_values('dec') == 'dec10']
    dec10.index = dec10.index.droplevel(level=0)
    dec10.name = 'long'
    dec1 = holdmean.loc[holdmean.index.get_level_values('dec') == 'dec1']
    dec1.index = dec1.index.droplevel(level=0)
    dec1.name = 'short'
    momret = pd.concat([dec10, dec1], axis=1)
    momret['mom'] = momret.long - momret.short
    return momret.mom
```
