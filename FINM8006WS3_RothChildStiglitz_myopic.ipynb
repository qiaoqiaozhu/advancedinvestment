## Rothschild and Stiglitz Theorem
Outcome $\tilde{X}$ is weakly less risky than outcome $\tilde{Y}$ if $\tilde{Y}=\tilde{X}+\tilde{\epsilon}$, where
$E(\tilde{\epsilon}|X)=0$. That is, $E[U(\tilde{Y})]\leq E[U(\tilde{X})]$.

### Proof:

Taking conditional expectations and using Jensen's inequality give
$$E[U(\tilde{Y}|X)]=E[U(\tilde{X}+\tilde{\epsilon})|X]\leq U[E(\tilde{X}+\tilde{\epsilon}|X)]=U[E(\tilde{X}|X)]=U(X)$$
Now taking expectation of conditional expectations over all $\tilde{X}$ values gives
$$ E(E[U(\tilde{Y}|X)])\equiv E[U(\tilde{Y})] \leq E[U(\tilde{X})]$$. The first $\equiv$ is called law of iterated expectations.

### Application:  Diversification of IID Payoffs

Consider $n$ lotteries with payoffs $\tilde{x_{1}}, \tilde{x_{1}},\dots, \tilde{x_{n}},$ that are independent and identically distributed (iid). You are asked to choose weights $\alpha_{1}, \alpha_{2},\dots, \alpha_{n}$ subject to the constraint that $\sum_{i} \alpha_{i} = 1$. It seems obvious that the best choice is complete diversification, with weights $\alpha_{i} = 1/n$ for all $i$. The payoff is then
 $$\tilde{z}=\frac{1}{n}\sum_{i=1}^{n}\tilde{x_{i}} $$
Let's prove it using the theorem. 

the payoff on any other strategy, $a_{1}, a_{2},\dots, a_{n}$ is:
$$\sum_{i=1}^{n}a_{i}\tilde{x_{i}}= \tilde{z}+\sum_{i=1}^{n}(a_{i}-\frac{1}{n})\tilde{x_{i}}=\tilde{z}+\tilde{\epsilon} $$
 where
$$ \tilde{\epsilon} \equiv \sum_{i=1}^{n}(a_{i}-\frac{1}{n})\tilde{x_{i}}$$
so that 
$$ E[\tilde{\epsilon}|z]=\sum_{i=1}^{n}(a_{i}-\frac{1}{n})E[\tilde{x_{i}}|z]$$

Since $\tilde{x_{i}}$ is iid, the conditional expectation must be some identical number, let's say k,  
$$ E[\tilde{\epsilon}|z] =k\sum_{i}(a_{i}-\frac{1}{n})=0$$
$\tilde{\epsilon}$ is pure noise. $\tilde{z}$ is preferred to all other allocation, and therefore optimal. 

### A Simple Example

Suppose under cdf F() the outcome of a game is either \$1 or \$3 with equal probability. Now suppose whenever the outcome is \$3 we add a dollar with probability $1/2$. let G denote the distribution of the outcome of the compound lottery. Quite clearly anyone with nondecreasing utility function would prefer the second compound lottery G.


```python
def F(x):
    if x< 1:
        return 0
    if 1 <=x < 3:
        return 0.5
    elif x>=3:
        return 1

def G(x):
    if x < 1:
        return 0
    if 1 <=x < 3:
        return 0.5
    elif 3<=x<4:
        return 0.75
    elif x>=4:
        return 1 
```

```python
x = np.linspace(0,4,50)
y = [F(xs) for xs in np.nditer(x)]
z = [G(xs) for xs in np.nditer(x)]
fig = plt.figure()
plt.plot(x,y, label=r'$L_F$')
plt.plot(x,z, label=r'$L_G$')
plt.legend()
plt.title('Lottory CDFs')
plt.show()
```

Now suppose whenever the outcome is 3, we add or subtract a dollar with equal probability. So the resulting distribution is a MPS. Draw CDF F and G, is there stochastic dominance?


```python
def newG(x):
    if x < 1:
        return 0
    if 1 <=x < 2:
        return 0.5
    elif 2<=x<4:
        return 0.75
    elif x>=4:
        return 1 
    
x = np.linspace(0,4,50)
y = [F(xs) for xs in np.nditer(x)]
z = [newG(xs) for xs in np.nditer(x)]
fig = plt.figure()
plt.plot(x,y, label=r'$L_F$')
plt.plot(x,z, label=r'$L_G$')
plt.legend()
plt.title('Lottory CDFs')
plt.show()
```

### Lognormal Distribution Property
If X is lognormal distributed, i.e.
$ log(X) \sim N(\mu, \sigma^{2}) $
then
$$ E[X] = e^{\mu+\frac{1}{2}\sigma^{2}} $$
or

$$\log E[X]= E [\log X] + \frac{1}{2} Var(\log X) =\mu+\frac{1}{2}\sigma^{2} $$

### SDF Pricing with Consumption
Lognormal Example

Let's consider a two period represent consumer economy with lognormal consumption growth. $c_{1}(z)/c_{0}=e^{z}$, where $z\sim N(\mu, \sigma^2)$. $$\Delta \log(c)=\log c_{1}- \log c_{0} =z$$, $$U=\dfrac{c^{1-\gamma}}{1-\gamma}$$

#### SDF 
$$m=\beta\frac{u'(c_{1})}{u'(c_{0})} $$
$$=\beta (\frac{c_{1}}{c_{0}})^{-\gamma}$$

$$ \log(m) = \log\beta - \gamma z $$

Clearly, m is lower when $z$ is higher. Good states have lower value. The strength of the effect depends on $\gamma$. 

$$\log(m) \sim N(\log \beta -\gamma \mu, \gamma^{2}\sigma^{2}) $$

Using the lognormal property, 
$$ E(m) = e^{\log \beta -\gamma \mu+ \frac{1}{2} \gamma^{2}\sigma^{2}} $$

$$ r_{f} = \frac{1}{E(m)} = \frac{1}{\beta} e^{\gamma \mu - \frac{1}{2}\gamma^{2}\sigma{2}} $$

It's remarkable we now have a risk-free rate just by constructing very simple models. However, the number seems a bit too big. It's also not just because consumption volatility is small - increasing $\sigma$ has small effect on $r_f$ unless by very unrealistic amount. (That's because variance of growth will be small anyway, see $log(R_f)$.)

This is still a very active research area. People has been modifying utility function; decomposing consumption growth into permanent and transitory; splitting growth into predictable and changing volatility (long-run risk), etc.  

We can decouple m with consumption, but how volatile sdf needs to be?
$$E(R^{i})-R^{f}=-R^{f}cov(m,R^{i}) =-R^{f}\rho \sigma_{i}\sigma_{m} $$

### Myopic Optimization with CARA

$$\max E_{t} [-e^{-\gamma W_{t+1}}] $$ subject to
$$ W_{t+1}=(1+R_{p, t+1})W_{t}$$ where the portfolio return is
$$ R_{p, t+1}=\alpha_{t}R_{t+1}+(1-\alpha_{t})R_{f,t+1}=R_{f,t+1}+ \alpha_{t}(R_{t+1}-R_{f,t+1})$$ 
Try to verify your answer

Apply lognormal to the objective function, where $var(-\gamma W_{t+1})=\gamma^{2}\alpha^{2}\sigma^2$
$$ \max E_{t}[ -\gamma W_{t+1} + \frac{1}{2} \gamma^{2}\alpha^{2}\sigma^2 ]$$

$$ -\gamma (1+R_{f,t+1}+ \alpha_{t}(R_{t+1}-R_{f,t+1}))W_{t} + \frac{1}{2}\alpha^{2} \gamma^{2}\sigma^2$$

FOC wrt $\alpha$
$$ \gamma\sigma^{2} \alpha =  (R_{t+1}-R_{f,t+1})W_{t} $$
$$  \alpha =  \frac{(R_{t+1}-R_{f,t+1})W_{t}}{\gamma\sigma^{2}}$$
