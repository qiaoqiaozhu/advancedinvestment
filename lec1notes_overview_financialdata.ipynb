Back to S&P 500. What if we base our information on lag one-day return? Is the conditional distribution normal? 

```python
def spcond(indata, retname, high=-0.01, low=-0.02):
    data = indata
    datalag = data.shift()
    lagname = 'lag'+ retname
    datalag.name = lagname
    data = pd.concat([data, datalag], 1)
    data = data[(data[lagname]<=high)&(data[lagname]>=low)]
    data = data[retname].dropna().as_matrix()
    # Fit a normal distribution to the data:
    mu, std = stats.norm.fit(data)
    k = stats.kurtosis(data)
    s = stats.skew(data)
    # Plot the histogram.
    fig, ax = plt.subplots()
    plt.hist(data, bins=250, normed=True, alpha=0.4, color='b')
    # Plot the PDF.
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 250)
    p = stats.norm.pdf(x, mu, std)
    plt.plot(x, p, 'r', linewidth=2)
    if retname == 'spret':
        title = "S&P Daily Returns Conditional on %.2f <= Lag Return <= %.2f \n mu = %.4f,\
        std = %.4f, Skewness = %.2f, Kurtosis = %.2f" % (low, high, mu, std, s, k)
    if retname == 'shret':
        title = "Shanghai Exchange Index  Daily Returns Conditional on %.2f <= Lag Return <= %.2f \n mu = %.4f,\
        std = %.4f, Skewness = %.2f, Kurtosis = %.2f" % (low, high, mu, std, s, k)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    plt.title(title)
    plt.show()
```

#### Conditional distribution

The concept of conditional distribution is the distribution (of stock returns) based on known information, for example, return yesterday. 

```python
# conditional distribution based on lag one-day return for s&P 500
# negative between roughly one and two std
spcond(sp['spret'],'spret', high = -0.01, low = -0.02)
```

#### Monthly Frenquency

```python
def histplot(var, series='S&P 500'):
    data = var.dropna().as_matrix()
    # Fit a normal distribution to the data:
    mu, std = stats.norm.fit(data)
    k = stats.kurtosis(data)
    s = stats.skew(data)
    # Plot the histogram.
    fig, ax = plt.subplots()
    plt.hist(data, bins=100, normed=True, alpha=0.5, color='g')
    # Plot the PDF.
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 250)
    p = stats.norm.pdf(x, mu, std)
    plt.plot(x, p, 'r', linewidth=2)
    title = "%s Monthly Returns mu = %.4f, std = %.4f, Skewness = %.2f, Kurtosis = %.2f" %(series, mu, std, s, k)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    plt.title(title)
    plt.show()
```

### Two main pricing approaches
* Our model should incorporate these stylized facts about return data
 - across asset classes
 - once they are grouped into baskets
 - in relation to the macroeconomy
 - depending on firm characteristics or prior performance
* If we try to price assets in reference to exposure to fundamental risk (e.g., GDP), then it's *absolute pricing*. It's hard but important. 
* If we try to price assets relative to prices of other assets, piggybacking on existing price observations (arbitrage), then it's *relative pricing*. It's easier but not without caveats.
